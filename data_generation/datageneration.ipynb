```python
import numpy as np
import pandas as pd
import os
import random
import json
```

```python


operationlist = ['Str','Dry','Heat','Warm','Distill','Coll','Held','Charge']
durationpre = ['Hour','Min','Second']
material = ['Tetrahydrofuran','Anhydrous','Propionic Acid Potassium', 'M-butyl Phospine HBF4','PF-06932437','PF-0631281']
length_state = 30
re = {}
re['statisticalInfo'] = {}
re['fileInfo'] = []
for i in range(10):
    operation = np.random.choice(operationlist,length_state)
    duration = [str(i)+' ' + j for i,j in zip(np.random.choice(range(1,10),length_state),np.random.choice(durationpre,length_state))]
    temperature = list(np.random.choice(range(5,110),length_state))
    material = list(np.random.choice(material,3))

    datadict = {'operation':operation, 'duration':duration,'temperature':temperature}
    df = pd.DataFrame(datadict)

    df.index.name = 'id'
    tempdata = df.to_json(orient='columns')
    tempdata = json.loads(tempdata)


    fakefile = {}
    fakefile['filename'] = 'record' + str(i)
    fakefile['fileid'] = str(i)
    fakefile['tempdata'] = tempdata
    fakefile['material'] = material
    
    re['fileInfo'].append(fakefile)
with open('datatotal' '.json','w') as f:
    json.dump(re,f)
```
